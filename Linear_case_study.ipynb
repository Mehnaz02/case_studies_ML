```python
data_multi=normal_d.drop(['carheight','compressionratio','boreratio','stroke','curbweight','peakrpm','enginesize'],axis=1)
data_multi
```

```python
data_dummies=pd.get_dummies(data_multi,drop_first=True)
data_dummies
```

```python
#Linear_Regresiion
import statsmodels.api as sm

x1=data_dummies.drop(['log_price'],axis=1)
x=sm.add_constant(pd.DataFrame(x1))
```

```python
y=data_dummies['log_price']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
sc.fit(x)
```

```python
input_scaler=sc.transform(x)
```

```python
input_scaler
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(input_scaler, y, test_size = 0.2, random_state=42)
```

```python
result=sm.OLS(y_train,x_train).fit()
```

```python
result.summary()
```

```python
y_pred=result.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_error
lin_mae = mean_absolute_error(y_test,y_pred)
lin_mae
```
