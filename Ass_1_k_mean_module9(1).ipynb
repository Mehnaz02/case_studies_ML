```python
import pandas as pd 
import numpy as np
```

```python
data=pd.read_csv('/home/mehnaz/Downloads/customers.csv')
```

```python
data.isna().sum()
```

```python
import matplotlib.pyplot as plt
plt.scatter(x=data['Age'],y=data['Spending Score (1-100)'])
```

```python
from sklearn.cluster import KMeans
```

```python
f1=data['Age'].values
f2=data['Spending Score (1-100)'].values
x=np.array(list(zip(f1,f2)))
```

```python
model=KMeans(n_clusters=2)
model.fit(x)
labels=model.predict(x)
```

```python
data['labels']=labels
```

```python
plt.scatter(data['Age'],data['Spending Score (1-100)'],c=data['labels'])
```
