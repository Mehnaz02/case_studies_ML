```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_breast_cancer
```

```python
bc=load_breast_cancer()
```

```python
bc.keys()
```

```python
data=pd.DataFrame(bc.data,columns=bc.feature_names)
```

```python
#lets normalize data
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
scaled=sc.fit_transform(data)
```

```python
pd.DataFrame(scaled.T)
```

```python
Tsc=scaled.T
```

```python
# now we find the coveraince

cov_mat=np.cov(Tsc)
```

```python
cov_mat
```

```python
#now after coverance we find eig_val, eig_vec

eig_val,eig_vec=np.linalg.eig(cov_mat)
```

```python
eig_val
```

```python
eig_vec
```

```python
# now we have to perform explained variance
total=sum(eig_val)
exp_var=[(i/total)*100 for i in sorted(eig_val,reverse=True)]
```

```python
exp_var
```

```python
cum_f=np.cumsum(exp_var)
```

```python
pd.DataFrame(cum_f)
```

```python
#scree plot
plt.figure(figsize=(12,6))
plt.bar(range(len(exp_var)),exp_var,color='purple')
plt.step(range(len(cum_f)),cum_f,color='yellow')
```
